```python
!pip install PyStasm
import stasm
import os
import cv2
from google.colab.patches import cv2_imshow
```

```python
data_path = '/content/drive/MyDrive/poc/faces'
img_path = os.path.join(data_path, '00011.png')
img = cv2.imread(img_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img[::4, ::4])
```

```python
landmarks = stasm.search_single(gray)
```

```python
img_copy = img.copy()
for pt in landmarks:
  cv2.circle(img_copy, (pt[0], pt[1]), 5, (255, 0, 0), -1)
  
cv2_imshow(img_copy)
```

```python
bottom = 49
upper = [17, 24]
left = 0
right = 12

y1 = min(int(landmarks[upper[0]][1]), int(landmarks[upper[1]][1])) - 10 #marge de 10 pixels
y2 = int(landmarks[bottom][1])

x1 = int(landmarks[left][0])
x2 = int(landmarks[right][0])

crop_img = img[y1:y2, x1:x2].copy()
```

```python
cv2_imshow(crop_img)
```
